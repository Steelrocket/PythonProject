```python
char = []
num = []
for i in train.columns:
    if train[i].dtype == 'O':
        char.append(i)
    else:
        num.append(i)
#also can do this
#train_num = train.select_dtypes(include=['number'])
#train_num.head()
```

```python
train_num = train[num]
```

```python
train_char = train[char]
```

```python
plt.figure(figsize=(12, 6))
train_num.isnull().sum().plot.bar()
plt.title("Numerical Data")
plt.xlabel('Numerical Columns')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(13,6))
train_char.isnull().sum().plot.bar()
plt.title("Qualitative Data")
plt.xlabel('Object Columns')
plt.ylabel('Count')
plt.show()
```
